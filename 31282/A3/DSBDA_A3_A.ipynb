```python
import pandas as pd

# Create a DataFrame with the provided data
df = pd.read_csv('/home/TE/Downloads/StudentsPerformance.xls')
```

```python

# Group by 'gender' and calculate summary statistics
summary_stats = df.groupby('race/ethnicity').describe().stack(level=0)

# Extract relevant columns for mean, median, min, max, and std
result = summary_stats[['mean', '50%', 'min', 'max', 'std']]

# Rename the columns for clarity
# result.columns = ['Mean', 'Median', 'Minimum', 'Maximum', 'Standard Deviation']

# Display the result
print(result)
```
