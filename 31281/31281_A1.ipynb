```python
data['Car'].fillna(0, inplace=True)
data['BuildingArea'].fillna(data['BuildingArea'].mean(), inplace=True)
data['YearBuilt'].fillna(data['YearBuilt'].mean(), inplace=True)
data['CouncilArea'].fillna('unknown', inplace=True)
```

```python
data.isnull().sum()
```

```python
data['Postcode'] = data['Postcode'].astype(int)
data['Bathroom'] = data['Bathroom'].astype(int)
data['Bedroom2'] = data['Bedroom2'].astype(int)
data['Car'] = data['Car'].astype(int)
data['YearBuilt'] = data['YearBuilt'].astype(int)
```

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
data['CouncilArea'] = label_encoder.fit_transform(data['CouncilArea'])
data
```
